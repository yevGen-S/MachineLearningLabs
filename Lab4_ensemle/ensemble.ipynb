# Задание 1

- Исследуйте зависимость качества классификации от количества классификаторов в ансамбле для алгоритмов бэггинга на наборе данных glass.csv с различными базовыми классификаторами. Постройте графики зависимости качества классификации при различном числе классификаторов, объясните полученные результаты.

```python
data = read_csv('glass.csv', sep=',').to_numpy()
data
```

```python
Y = data[:, -1]
X = np.delete(data, [len(data[0]) - 1, 0], axis=1)
X
```

```python
x_train, x_test, y_train, y_test = train_test_split(
   X, Y, test_size=0.33)
```

```python
n_s = [i for i in range(1, 200)]


def get_predictions(est, classifier=BaggingClassifier):
    predictions = []

    for n in n_s:
    clf = classifier(base_estimator=est,
                          n_estimators=n, random_state=0).fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    predictions.append(accuracy_score(y_test, y_pred))

    return predictions
```

```python
fig, sub = plt.subplots(2, 2, figsize=(15, 10))
plt.subplots_adjust(wspace=.3, hspace=.2)

classifiers = [
               SVC(),
               KNeighborsClassifier(),
               GaussianNB(),
               DecisionTreeClassifier()
]


for clf, ax in zip(classifiers, sub.flatten()):
    preds = get_predictions(clf)
    ax.plot(n_s, preds)
    ax.set_xlabel('number of estimators')
    ax.set_ylabel('accuracy')
    ax.set_title(clf.__class__.__name__)

plt.show()
```

# Задание 3

- Постройте мета-классификатор для набора данных titanic_train.csv используя стекинг и оцените качество классификации на titanic_train.csv

```python
from sklearn.metrics import confusion_matrix
import math
import sklearn.preprocessing as preprocessing
import itertools
```

```python
def prepare_dataset(filename, sep=','):
    data = read_csv(filename, sep=sep)
    data = data.astype(str)

    le = preprocessing.LabelEncoder()
    return data.apply(le.fit_transform)
```

```python
data = prepare_dataset('titanic.csv').to_numpy()
data = data.astype(str)
Y = data[:, 1]
X = np.delete(data, [0, 1], axis=1)
X = X.astype(int)
Y = Y.astype(int)

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
x_train
```

```python
classifiers = [
               ('svc', SVC()),
  ('k neighbours', KNeighborsClassifier()),
  ('naive bayes',GaussianNB()),
  ('decision tree', DecisionTreeClassifier())
]

clf = StackingClassifier(estimators=classifiers)
clf.fit(x_train, y_train)

confusion_matrix(y_test, clf.predict(x_test))
```

```python
for combo_num in range(1, len(classifiers) + 1):
    combos = np.array(list(itertools.combinations(classifiers, combo_num)))
  
    for i in combos:
        clf = StackingClassifier(estimators=classifiers)
        clf.fit(x_train, y_train)

    y_pred = clf.predict(x_test)
    
    classes_str = '+'.join('"' + i[:, 0] + '"')
    print('{} = \n{}\n'.format(classes_str, confusion_matrix(y_test, y_pred)))
```
