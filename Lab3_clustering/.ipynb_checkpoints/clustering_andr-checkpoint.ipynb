# Задание 1
 - Разбейте множество объектов из набора данных pluton.csv на 3 кластера с помощью k-means. Сравните качество разбиения в зависимости от максимального числа итераций алгоритма и использования стандартизации.

```python
X = read_csv('pluton.csv', sep=',').to_numpy()
```

```python
def visualize_data(clust, X,title, legend=True):
    clustering = clust.fit(X)
    plt.figure(figsize=(5, 5))

    labels = clustering.labels_.reshape(-1, 1)

    labels_set = list(set(clustering.labels_))

    all_data = np.concatenate((X, labels), axis=1)

    clusters = []

    for i in labels_set:
        clusters.append(all_data[all_data[:, -1] == i])

    clusters = np.array(clusters)

    for c in clusters:
        plt.scatter(c[:, 0], c[:, 1])

#     plt.xlabel('X1')
#     plt.ylabel('X2')

    if legend:
        plt.legend(['class %d' % i for i in labels_set])

    plt.grid(True)

    try:
        print(davies_bouldin_score(X, clustering.labels_))
    except Exception as e:
        print(e)
    plt.title(title)
    plt.show()
```

```python
def standardize(data):
    return StandardScaler().fit(data).transform(data)
```

```python
max_iters = np.array([1, 10, 100, 300, 1000, 2000])

Y = standardize(X)

for mi in max_iters:
    kmeans = KMeans(n_clusters=3, n_init=10, max_iter=mi, random_state=21).fit(X)
    visualize_data(kmeans, X,f'max_iter={mi}', legend=True)
    kmeans = KMeans(n_clusters=3, n_init=10, max_iter=mi, random_state=21).fit(Y)
    visualize_data(kmeans, Y,f'Standardize: max_iter={mi}', legend=True)
```

```python
n_clusters_arr = [i for i in range(2, 15)]
dbi_kmeans_arr = []
dbi_dbscan_arr = []
dbi_hierarcy_arr = []


from IPython.display import clear_output

X = X3

for n in n_clusters_arr:
    kmeans = KMeans(n_clusters=n, n_init=10).fit(X)
    aggclust = AgglomerativeClustering(n_clusters=n).fit(X)

    dbi_kmeans = davies_bouldin_score(X, kmeans.labels_)
    dbi_hierarcy = davies_bouldin_score(X, aggclust.labels_)
    
    dbi_kmeans_arr.append(dbi_kmeans)
    dbi_hierarcy_arr.append(dbi_hierarcy)
```

- Осуществите сжатие цветовой палитры изображения (любого, на ваш выбор). Для этого выделите n кластеров из цветов всех пикселей изображения и зафиксируйте центра этих кластеров. Создайте изображение с цветами из сокращенной палитры (цвета пикселей только из центров выделенных кластеров). Покажите исходное и сжатое изображения.

```python
from skimage import io
```

```python
def compress_image(image, clusters_num):
    rows = image.shape[0]
    cols = image.shape[1]

    image = image.reshape(rows*cols, 3)

    kmeans = KMeans(n_clusters=clusters_num, n_init=10)
    kmeans.fit(image)
    compressed_image = kmeans.cluster_centers_[kmeans.labels_]
    compressed_image = np.clip(compressed_image.astype('uint8'), 0, 255)
    compressed_image = compressed_image.reshape(rows, cols, 3)
    io.imsave('compressed_image_%d.png' % clusters_num, compressed_image)
```

```python
clusters_nums = [64, 32, 16, 8]
for cn in clusters_nums:
    compress_image(io.imread('elephant.jpeg'), cn)
    print(cn, 'done')
```

# Задание 4

- Постройте дендрограмму для набора данных votes.csv (число голосов, поданных за республиканцев на выборах с 1856 по 1976 год). Строки представляют 50 штатов, а столбцы - годы выборов (31). Проинтерпретируйте полученный результат.

```python
from scipy.cluster.hierarchy import dendrogram


def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    dendrogram(linkage_matrix, **kwargs)
```

```python
X = read_csv('votes.csv', sep=',').to_numpy()
X = np.nan_to_num(X, nan=0)
X
```

```python
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
model = model.fit(X)

plt.figure(figsize=(15, 15))
plt.grid(True)
plt.title('Hierarchical Clustering Dendrogram')
plot_dendrogram(model, truncate_mode='level', p=15)
plt.show()
```
